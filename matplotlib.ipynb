```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(5)
y = np.random.normal(120, 7, 30)
x = [i for i in range(1970, 2000)]
```

```python
plt.plot(x, y)
# plt.clf()
plt.scatter(x,y)
plt.title('RANDOM BULLSHIT')
plt.xlabel('random bs')
plt.ylabel('random bs 2')
```
